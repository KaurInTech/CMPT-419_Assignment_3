```python
import random
from datasets import load_dataset
import pandas as pd
```

```python
num_files = 5535

num_files_to_sample = num_files * 0.001
num_files_to_sample
```

```python
file_numbers = random.sample(range(0,num_files), int(num_files_to_sample))
```

```python
#%%
# just a quick reminder of what the number formatting looks like
number = 35
f"{number:05d}"
```

```python
#%%
pattern = "data/train-{:05d}-of-05534-*"
file_patterns = [pattern.format(x) for x in file_numbers]
file_patterns
```

```python
#%%
data_sample = load_dataset(
    "tiiuae/falcon-refinedweb",
    data_files=file_patterns, verification_mode='no_checks'
)
df = pd.DataFrame(data_sample["train"])
```

```python
#%%
name = '-'.join([str(x) for x in file_numbers])
df.to_parquet(f'refinedweb_{name}.parquet')
```

```python
import tldextract
```

```python
# === GET DOMAIN AND SUBDOMAIN ===
domains = []
subdomains = []
full = []
for url in df.url.values:
    extract = tldextract.extract(url)
    domains.append(extract.domain + '.' + extract.suffix)
    subdomains.append(extract.subdomain)
    full.append(extract.subdomain + '.' + extract.domain + '.' + extract.suffix)
df["domain"] = domains
df["subdomain"] = subdomains
df["full"] = full
df['totaldocuments'] = 1
```

```python
#%%
# === COUNT WORDS === 
df['totalwords'] = [len(x.split()) for x in df['content'].tolist()]
```

## Dataset 2: Dolmo

### Data was extracted using git clone:
### git clone https://huggingface.co/datasets/allenai/dolma dolma
### shuf -n 6 "dolma/urls/v1_6.txt" | xargs -n 1 -P "6" wget -q -P data

```python
import os

# Directory containing the JSON files
directory = "data"

# List to store file names
file_names = []

# Iterate over files in the directory
for file_name in os.listdir(directory):
    # Check if the file is a JSON file
    if file_name.endswith(".json.gz"):
        # Add the file name to the list
        file_names.append(os.path.join(directory, file_name))

for file_name in file_names:
    print(file_name)
```

```python
import pandas as pd
import gzip

# List to store DataFrames
dfs = []

# Number of rows to read from each file as the files are very long
num_rows_per_file = 800

# Iterate over file names in the list
for filename in file_names:
    # Open the compressed JSON file and read its contents
    with gzip.open(filename, "rb") as f:
        # Read the decompressed JSON data into a DataFrame
        df = pd.read_json(f, lines=True, nrows=num_rows_per_file)
        # Append the DataFrame to the list
        dfs.append(df)

# Concatenate all DataFrames into a single DataFrame
combined_df = pd.concat(dfs, ignore_index=True)
```

```python
combined_df
```

```python
# Sample 50 random observations
combined_df_sampled = combined_df.sample(n=50, random_state=42)

# Add empty column named "observations"
combined_df_sampled['observations'] = ''

# Write sampled data to CSV file
combined_df_sampled.to_csv('dolma_sample.csv', index=False, na_rep='')

# Read CSV file
dolma_sample = pd.read_csv('dolma_sample.csv')

dolma_sample
```
