```python
import random
from datasets import load_dataset
import pandas as pd
```

```python
num_files = 5535

num_files_to_sample = num_files * 0.001
num_files_to_sample
```

```python
file_numbers = random.sample(range(0,num_files), int(num_files_to_sample))
file_numbers
```

```python
#%%
# just a quick reminder of what the number formatting looks like
number = 35
f"{number:05d}"
```

```python
#%%
pattern = "data/train-{:05d}-of-05534-*"
file_patterns = [pattern.format(x) for x in file_numbers]
file_patterns
```

```python
#%%
data_sample = load_dataset(
    "tiiuae/falcon-refinedweb",
    data_files=file_patterns, verification_mode='no_checks'
)
df = pd.DataFrame(data_sample["train"])
```

```python
#%%
name = '-'.join([str(x) for x in file_numbers])
df.to_parquet(f'refinedweb_{name}.parquet')
```

```python
# To use saved file, specify the name here
name = '1987-2854-2843-3453-817'
path = f'refinedweb_{name}.parquet'

df = pd.read_parquet(path)
```

```python
import os

# Get the size of the Parquet file in bytes
file_size_bytes = os.path.getsize(path)

# Convert bytes to megabytes (MB)
file_size_mb = file_size_bytes / (1024 * 1024)

print(f"Size of the Parquet file: {file_size_mb:.2f} MB")
```

```python
# Sample 50 random observations
df_sampled = df.sample(n=50, random_state=42)  # Adjust random_state for reproducibility

# Add empty column named "observations"
df_sampled['observations'] = ''

# Write sampled data to CSV file
df_sampled.to_csv('refinedweb_sample.csv', index=False, na_rep='')

# Read CSV file
refinedweb_sample = pd.read_csv('refinedweb_sample.csv')

refinedweb_sample
```

```python
import tldextract
```

```python
# === GET DOMAIN AND SUBDOMAIN ===
domains = []
subdomains = []
full = []
for url in df.url.values:
    extract = tldextract.extract(url)
    domains.append(extract.domain + '.' + extract.suffix)
    subdomains.append(extract.subdomain)
    full.append(extract.subdomain + '.' + extract.domain + '.' + extract.suffix)
df["domain"] = domains
df["subdomain"] = subdomains
df["full"] = full
df['totaldocuments'] = 1
```

```python
#%%
# === COUNT WORDS === 
df['totalwords'] = [len(x.split()) for x in df['content'].tolist()]
```
